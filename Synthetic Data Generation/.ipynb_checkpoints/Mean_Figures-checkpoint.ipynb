```python
from fvgp.gp import GP

import torch
from torch import nn

import numpy as np
from numpy.random import default_rng

import matplotlib.pyplot as plt
import plotly.graph_objects as go

import os
import csv

import random
```

# Loading Data

```python
energy_data = np.load("C:\\Users\\Maher Alghalayini\\Desktop\\Paper 1 Material\\Synthetic Data Generation\\my_synthetic_energy.npy")
cycle_number = np.load("C:\\Users\\Maher Alghalayini\\Desktop\\Paper 1 Material\\Synthetic Data Generation\\my_synthetic_cycleNum.npy")


num_of_datasets = len(energy_data[:,0])

label_size = 30
plt.figure(figsize = (20,10))
for i in range(len(energy_data)): plt.scatter(cycle_number,energy_data[i])

plt.tick_params(axis='both', which='major', labelsize=label_size) # Set the font size of the tick labels on the x and y axes
plt.xlabel("Cycle Number",fontsize=label_size)
plt.ylabel("Energy",fontsize=label_size)
plt.show()

print("max y: ", np.max(energy_data))

# Initializing the data to fit the GP model
data_size = num_of_datasets

# All Data
x_data = np.tile(cycle_number, data_size).reshape(-1, 1) # repeat cycle 20 times to create x_data
y_data = np.vstack(energy_data).reshape(-1, 1) # stack energy rows to create y_data
x_pred = np.linspace(0,1000,1001).reshape(-1,1)


print("x data: ", x_data.shape)
print("y data: ", y_data.shape)
        
```

# 1 - Power-law Model

## GP Components 

```python
def get_distance_matrix(x1,x2):
    d = np.zeros((len(x1),len(x2)))
    for i in range(x1.shape[1]):
        d += (x1[:,i].reshape(-1, 1) - x2[:,i])**2
    return np.sqrt(d)

# For the Noise
def s(x, my_slope, my_pow, my_intercept):
    o = my_slope * (x**my_pow) + my_intercept
    return o


def my_noise(x,hps,obj):

    my_slope     = hps[2]
    my_pow       = hps[3]
    my_intercept = hps[4]
    
    noise = np.identity(len(x)) * s(x,my_slope,my_pow,my_intercept)

    return noise

def kernel(x1,x2,hps,obj):
    d = get_distance_matrix(x1,x2) 
    k = hps[0] * obj.squared_exponential_kernel(d,hps[1])
    return k


def mean(x,hps,obj):
    return -( hps[5] * x[:,0])**2 + hps[6]
```

## GP Modeling

```python
# Initializing the GP Model
init_hyperparameters = np.array([1500, 500,              # Kernel
                                 0.05, 0.48, 52,         # Noise
                                 1, 500])                # Mean

my_gp = GP(1, x_data, y_data,init_hyperparameters,
            gp_kernel_function=kernel, gp_noise_function=my_noise,gp_mean_function = mean)


# Setting the Optimization Bounds for Hyperparameters
bounds = np.empty((7,2))

# Kernel Sq Exp
bounds[0] = np.array([1e-5,6000.])                     # Kernel Variance
bounds[1] = np.array([100.,1000.])                     # Kernel Lengthscale

# Noise
bounds[2] = np.array([1e-10,1.])                       # Noise Slope
bounds[3] = np.array([2,5.])                           # Noise Power
bounds[4] = np.array([0.,6.])                          # Noise Intercept
# Mean
bounds[5] = np.array([1e-5,100.])                      # Mean slope
bounds[6] = np.array([400.,600.])                       # Mean intercept

my_gp.train(bounds,max_iter=100)
```

```python
#print("hps: ", my_gpo.hyperparameters)
f = my_gp.posterior_mean(x_pred)["f(x)"]
v = my_gp.posterior_covariance(x_pred)["v(x)"]

# Plotting the data
plt.figure(figsize = (20,10))
plt.plot(x_pred[:,0],f, color = "blue", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black") # Training Data
plt.fill_between(x_pred[:,0],f - 2. * np.sqrt(v), f + 2. * np.sqrt(v), alpha = 0.5, color = "grey")
plt.xlabel("cycle number") 
plt.ylabel("energy")
plt.legend()
plt.show()
```

## Plotting the Prior Mean

```python
print("hps: ", my_gp.hyperparameters)

trained_hps = np.array( [1.42360706e+02, 4.44174830e+02, 1.05910771e-06, 2.89074057e+00,
 2.26137589e+00, 7.96447823e-03, 4.95705252e+02])


trained_prior_y = mean(x_pred,my_gp.hyperparameters,2)


# Plotting the data
plt.figure(figsize = (15,10))
#plt.plot(x_pred[:,0],prior_y, color = "red", linewidth = 6, label = "Trained Prior")
plt.plot(x_pred[:,0],trained_prior_y, color = "red", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black",label = "Experimental Data") # All Data
plt.tick_params(axis='both', which='major', labelsize=label_size) # Set the font size of the tick labels on the x and y axes
plt.xlabel("Cycle Number",fontsize=label_size) 
plt.ylabel("Energy",fontsize=label_size)
plt.legend(fontsize=label_size,frameon=False)
plt.show()
```

#  2 - Two-Element Piecewise Model

## GP Components

```python
def get_distance_matrix(x1,x2):
    d = np.zeros((len(x1),len(x2)))
    for i in range(x1.shape[1]):
        d += (x1[:,i].reshape(-1, 1) - x2[:,i])**2
    return np.sqrt(d)


# For the Noise
def s(x, my_slope, my_pow, my_intercept):
    o = my_slope * (x**my_pow) + my_intercept
    return o


def my_noise(x,hps,obj):

    my_slope     = hps[2]
    my_pow       = hps[3]
    my_intercept = hps[4]
    
    noise = np.identity(len(x)) * s(x,my_slope,my_pow,my_intercept)

    return noise

def kernel(x1,x2,hps,obj):
    d = get_distance_matrix(x1,x2) 
    k = hps[0] * obj.squared_exponential_kernel(d,hps[1])
    return k

# here I am assuming that the mean function is a piecewise function
def mean2(x,hps,obj):
    x0 = hps[5]

    m1 = hps[6]
    m2 = hps[7]

    b1 = 497
    b2 = (m1 - m2) * x0 + b1

    y = np.where(x[:,0] <= x0, m1*x[:,0] + b1, m2*x[:,0] + b2)
    return y
```

## GP Modeling

```python
# Initializing the GP Model
init_hyperparameters = np.array([1500, 500,              # Kernel
                                 0.05, 0.48, 52,         # Noise
                                 400, -1e-1,-1e-2])      # Mean

my_gp1 = GP(1, x_data, y_data,init_hyperparameters,
            gp_kernel_function=kernel, gp_noise_function=my_noise,gp_mean_function = mean2)


# Setting the Optimization Bounds for Hyperparameters
bounds = np.empty((8,2))

# Kernel Sq Exp
bounds[0] = np.array([1e-5,6000.])                     # Kernel Variance
bounds[1] = np.array([100.,1000.])                     # Kernel Lengthscale

# Noise
bounds[2] = np.array([1e-10,1.])                       # Noise Slope
bounds[3] = np.array([2,5.])                           # Noise Power
bounds[4] = np.array([0.,6.])                          # Noise Intercept
# Mean
bounds[5] = np.array([350.,750.])                      # Mean Elements Intersection
bounds[6] = np.array([-1e-1,-1e-3])                    # Mean Slope 1
bounds[7] = np.array([-1e-1,-1e-3])                     # Mean Slope 2

my_gp1.train(bounds,max_iter=100)
```

```python
#print("hps: ", my_gpo.hyperparameters)
f1 = my_gp1.posterior_mean(x_pred)["f(x)"]
v1 = my_gp1.posterior_covariance(x_pred)["v(x)"]

# Plotting the data
plt.figure(figsize = (20,10))
plt.plot(x_pred[:,0],f1, color = "blue", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black") # Training Data
plt.fill_between(x_pred[:,0],f1 - 2. * np.sqrt(v1), f1 + 2. * np.sqrt(v1), alpha = 0.5, color = "grey")
plt.xlabel("cycle number") 
plt.ylabel("energy")
plt.legend()
plt.show()
```

## Plotting the Prior Mean

```python
print("hps: ", my_gp1.hyperparameters)

trained_hps = np.array( [ 3.08259435e+00,  6.59542896e+02,  6.59126738e-07,  2.96729376e+00,
  2.05182610e+00,  4.66963250e+02, -2.71011471e-02, -9.74318979e-02])

trained_prior_y1 = mean2(x_pred,my_gp1.hyperparameters,2)


# Plotting the data
plt.figure(figsize = (15,10))
plt.plot(x_pred[:,0],trained_prior_y1, color = "red", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black",label = "Experimental Data") # All Data
plt.tick_params(axis='both', which='major', labelsize=label_size) # Set the font size of the tick labels on the x and y axes
plt.xlabel("Cycle Number",fontsize=label_size) 
plt.ylabel("Energy",fontsize=label_size)
plt.legend(fontsize=label_size,frameon=False)
plt.show()
```

# 3- Three-Element Piecewise Function

## GP Components

```python
def get_distance_matrix(x1,x2):
    d = np.zeros((len(x1),len(x2)))
    for i in range(x1.shape[1]):
        d += (x1[:,i].reshape(-1, 1) - x2[:,i])**2
    return np.sqrt(d)

# For the Noise
def s(x, my_slope, my_pow, my_intercept):
    o = my_slope * (x**my_pow) + my_intercept
    return o

def my_noise(x,hps,obj):
    my_slope     = hps[2]
    my_pow       = hps[3]
    my_intercept = hps[4]
    
    noise = np.identity(len(x)) * s(x,my_slope,my_pow,my_intercept)

    return noise

def kernel(x1,x2,hps,obj):
    d = get_distance_matrix(x1,x2) 
    k = hps[0] * obj.squared_exponential_kernel(d,hps[1])
    return k


def mean3(x,hps,obj):
    x0 = hps[5]
    x1 = x0 + hps[6]

    m1 = hps[7]
    m2 = hps[8]
    m3 = hps[9]

    b1 = 497
    b2 = (m1 - m2) * x0 + b1
    b3 = (m2 - m3) * x1 + b2

    y = np.where(x[:, 0] <= x0, m1*x[:, 0] + b1, 
                 np.where(x[:, 0] <= x1, m2*x[:, 0] + b2, 
                          m3*x[:, 0] + b3))
    return y
```

## GP Modeling

```python
# Initializing the GP Model
init_hyperparameters = np.array([1500, 500,                          # Kernel
                                 0.05, 0.48, 52,                     # Noise
                                 400, 100, -1e-1,-1e-2, -5e-2])      # Mean

my_gp2 = GP(1, x_data, y_data,init_hyperparameters,
            gp_kernel_function=kernel, gp_noise_function=my_noise,gp_mean_function = mean3)


# Setting the Optimization Bounds for Hyperparameters
bounds = np.empty((10,2))

# Kernel Sq Exp
bounds[0] = np.array([1e-5,6000.])                     # Kernel Variance
bounds[1] = np.array([100.,1000.])                     # Kernel Lengthscale

# Noise
bounds[2] = np.array([1e-10,1.])                       # Noise Slope
bounds[3] = np.array([2,5.])                           # Noise Power
bounds[4] = np.array([0.,6.])                          # Noise Intercept
# Mean
bounds[5] = np.array([350.,750.])                      # Mean Elements Intersection
bounds[6] = np.array([50.,400.])                       # Mean Elements Intersection
bounds[7] = np.array([-1e-1,-1e-3])                    # Mean Slope 1
bounds[8] = np.array([-1e-1,-1e-3])                    # Mean Slope 2
bounds[9] = np.array([-1e-1,-1e-3])                    # Mean Slope 2

my_gp2.train(bounds,max_iter=100)
```

```python
#print("hps: ", my_gpo.hyperparameters)
f2 = my_gp2.posterior_mean(x_pred)["f(x)"]
v2 = my_gp2.posterior_covariance(x_pred)["v(x)"]

# Plotting the data
plt.figure(figsize = (20,10))
plt.plot(x_pred[:,0],f2, color = "blue", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black") # Training Data
plt.fill_between(x_pred[:,0],f2 - 2. * np.sqrt(v2), f2 + 2. * np.sqrt(v2), alpha = 0.5, color = "grey")
plt.xlabel("cycle number") 
plt.ylabel("energy")
plt.legend()
plt.show()
```

## Plotting the Prior Mean

```python
print("hps: ", my_gp2.hyperparameters)

trained_hps = np.array( [ 1.08021769e+01,  8.73555229e+02,  1.88568842e-06,  2.80198045e+00,
  1.89328560e+00,  4.21691587e+02,  1.11041282e+02, -2.81935329e-02, -5.50047385e-02, -9.77729284e-02])

trained_prior_y2 = mean3(x_pred,my_gp2.hyperparameters,2)

# Plotting the data
plt.figure(figsize = (15,10))
plt.plot(x_pred[:,0],trained_prior_y2, color = "red", linewidth = 3, label = str(data_size) + " cells")
plt.scatter(x_data[:,0],y_data, color = "black",label = "Experimental Data") # All Data
plt.tick_params(axis='both', which='major', labelsize=label_size) # Set the font size of the tick labels on the x and y axes
plt.xlabel("Cycle Number",fontsize=label_size) 
plt.ylabel("Energy",fontsize=label_size)
plt.legend(fontsize=label_size,frameon=False)
plt.show()
```
