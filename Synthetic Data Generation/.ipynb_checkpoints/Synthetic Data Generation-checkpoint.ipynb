```python
from fvgp.gp import GP

import torch
from torch import nn

import numpy as np
from numpy.random import default_rng

import matplotlib.pyplot as plt
import plotly.graph_objects as go

import os
import csv

import random
```

```python
# Mean function: Two-Element piecewise function
def mean2(x,hps):

    x0 = hps[1]
    
    m1 = hps[2]
    m2 = hps[3] + m1

    b1  = hps[0]
    b2 = (m1 - m2) * x0 + b1

    y = np.where(x <= x0, m1*x + b1, m2*x + b2)
                
    return y
```

```python
x = np.linspace(0,1000,51)

hps = np.array([500,500,-0.05,-0.2])

y = mean2(x,hps)

plt.scatter(x,y)
```

# Generating the Synthetic Data

```python

# Generate random hyperparameters
arr = np.random.uniform(low=[495, 350, -8e-2, -1e-2], high=[500, 700, -3e-2, -1e-1], size=(20, 4))

label_size = 30

# Create a matrix to store prior_y values
prior_y_matrix = np.zeros((20, len(x)))

# Generate the data and populate the matrix
for i in range(20):
    prior_y = mean2(x, arr[i])
    prior_y_matrix[i, :] = prior_y

# Plotting the data
plt.figure(figsize=(15, 10))

for i in range(20):
    plt.plot(x, prior_y_matrix[i], color='0.65', linewidth=3)

plt.tick_params(axis='both', which='major', labelsize=label_size)
plt.ylim(380, 510)
plt.xlabel("Cycle Number", fontsize=label_size)
plt.ylabel("Energy", fontsize=label_size)
plt.show()
```

# Calculating the variance

```python
# Checking the variability trend in the data

variances = np.var(prior_y_matrix, axis=0)
stds = np.sqrt(variances)

#print(variances)

# Plotting the data
plt.figure(figsize = (15,10))
plt.plot(x,stds, color = "red", linewidth = 3)
plt.xlabel("Cycle Number",fontsize=label_size)
plt.ylabel("Energy Standard Deviation",fontsize=label_size)
plt.tick_params(axis='both', which='major', labelsize=label_size) # Set the font size of the tick labels on the x and y axes
plt.show()
```
